# Connect data set

```python
x_train = np.zeros((x_train1.shape[0] +x_train2.shape[0]+x_train3.shape[0]+x_train4.shape[0], 28, 28))
y_train = np.zeros((y_train1.shape[0] +y_train2.shape[0] +y_train3.shape[0] +y_train4.shape[0] , 10))
```

```python
idx = 0

for i in range(x_train1.shape[0]):
    x_train[idx] = x_train1[i]
    y_train[idx] = y_train1[i]
    
    idx += 1

for i in range(x_train2.shape[0]):
    x_train[idx] = x_train2[i]
    y_train[idx] = y_train2[i]
    
    idx += 1
    
for i in range(x_train3.shape[0]):
    x_train[idx] = x_train3[i]
    y_train[idx] = y_train3[i]
    
    idx += 1
    
    
for i in range(x_train4.shape[0]):
    x_train[idx] = x_train4[i]
    y_train[idx] = y_train4[i]
    
    idx += 1
```

# Reshape input

```python
x_train = np.reshape(x_train, (18515, 28, 28, 1))
```

## Build model cnn

```python
from keras.models import Sequential
from keras.layers import *
```

```python
model = Sequential()

model.add(InputLayer(input_shape=(28, 28, 1)))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))

model.add(Dense(128, activation='relu'))

model.add(Dense(10, activation='softmax'))
```

```python
epochs = 20
batch_size = 128
learning_rate = 0.01

loss = keras.losses.categorical_crossentropy

optimizer = keras.optimizers.Adam(lr=learning_rate)
```

```python
model.compile(
    loss=loss,
    optimizer=optimizer,
    metrics=['accuracy'])
```

```python
hist = model.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_split= 0.1)
```

```python
plt.figure(figsize=(10,6))
plt.plot(hist.history['accuracy'],label = 'acc')
plt.plot(hist.history['val_accuracy'],label = 'val_acc')
plt.legend()
```

```python
model.save('digit_model.h5')
```
